```python
import pandas as pd
```

```python
d1=pd.read_csv("real_temp.csv")
d2=pd.read_csv("fake_temp.csv")
```

```python
d1["label"]="real"
d2["label"]="fake"
```

```python
d3=pd.concat([d1,d2])
d3.columns
```

```python
d3=d3.drop(['Unnamed: 0'],axis=1)
```

```python
d4=d3.dropna(axis=0).reset_index(drop=True)
d4
```

```python
d4=d4.sample(frac=1).reset_index(drop=True)
d4
```

```python
d4.to_csv("all_news.csv")
```
