```python
data_real=pd.read_csv("real_main.csv",sep='|')
data_fake=pd.read_csv("fake_main.csv",sep='|')

data_real_meta=pd.read_csv("real_meta_cleaned.csv")
data_fake_meta=pd.read_csv("fake_meta_cleaned.csv")
```

```python
fc=data_fake_meta["follower_count"].groupby(data_fake_meta["news_item"])
rc=data_fake_meta["retweet_count"].groupby(data_fake_meta["news_item"])
fav_count=data_fake_meta["favorite_count"].groupby(data_fake_meta["news_item"])

print(fav_count.sum())
```

```python
data_fake.shape[0]
```

```python
list_follower_count=fc.sum()
dict_follower_count=list_follower_count.to_dict()





#for setting retweet count of each news
list_retweet_count=rc.sum()
dict_retweet_count=list_retweet_count.to_dict()

#for setting favourite count of each news
list_fav_count=fav_count.sum()
dict_fav_count=list_fav_count.to_dict()





data_fake["retweet_count"]=float("nan")
data_fake["favorite_count"]=float("nan")

data_fake.iloc[0]["retweet_count"]=100

k=data_fake.shape[0]
for i in range(0,k):
    news_item=data_fake.iloc[i]["news_item"]
    if(news_item in dict_retweet_count.keys()):
        data_fake.at[i,"retweet_count"]=dict_retweet_count[news_item]
        

for i in range(0,k):
    news_item=data_fake.iloc[i]["news_item"]
    if(news_item in dict_fav_count.keys()):
        data_fake.at[i,"favorite_count"]=dict_fav_count[news_item]








```

```python
#converting boolean columns to int for calculation.True is 1 and False is 0
data_fake_meta[['is_protected', 'is_verified']]=data_fake_meta[['is_protected', 'is_verified']].astype(int)
data_fake["is_verified_count"]=float("nan")

verified_count=data_fake_meta["is_verified"].groupby(data_fake_meta["news_item"])
list_verified_count=verified_count.sum()
dict_verified_count=list_verified_count.to_dict()

#data_fake["is_verified_count"]=float("nan")
for i in range(0,k):
    news_item=data_fake.iloc[i]["news_item"]
    if(news_item in dict_verified_count.keys()):
        data_fake.at[i,"is_verified_count"]=dict_verified_count[news_item]
        
data_fake.head()


```

```python
def datediffer(date1,date2):
    parser_str='%a %b %d %H:%M:%S +0000 %Y'
    return(datetime.strptime(date2,parser_str)-datetime.strptime(date1,parser_str)).days
```

```python

data_fake_meta["date_diff"]=0

for i in range(0,data_fake_meta.shape[0]):
    d1=data_fake_meta.iloc[i]["user_created_at"]
    d2=data_fake_meta.iloc[i]["tweet_created_at"]
    diff=datediffer(d1,d2)
    data_fake_meta.at[i,"date_diff"]=diff
    
```

```python
data_fake["mean_time_diff"]=float("nan")
mean_time_diff=data_fake_meta["date_diff"].groupby(data_fake_meta["news_item"])
list_mean_time_diff=mean_time_diff.mean()
dict_mean_time_diff=list_mean_time_diff.to_dict()

for i in range(0,k):
    news_item=data_fake.iloc[i]["news_item"]
    if(news_item in dict_mean_time_diff.keys()):
        data_fake.at[i,"mean_time_diff"]=dict_mean_time_diff[news_item]



```
